```python
taptap = 132.75
national_exchange = 133.70
amount = 1000 
step = 100
```

```python
X_taptap = [x for x in range(-1, amount, step)]
X_natinal_exchange = [x + 20 for x in range(-1, amount, step)]
Y_taptap = [taptap * x for x in range(-1, amount, step)]
Y_national_exchange = [national_exchange * (x - 5) if x < 500 else national_exchange * (x - 7) for x in range(-1, amount, step)]
```

```python
plt.bar(X_taptap, Y_taptap, width = 20, label='TapTap Send', color='blue')
plt.bar(X_natinal_exchange, Y_national_exchange, width = 20, label='National Exchange', color='limegreen')
plt.title('TapTap Send vs National Exchange')
plt.ylabel('BDT')
plt.xlabel('EURO')
plt.legend()
```
